```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast

df = pd.read_csv('tmdb_5000_movies.csv')

print(df)
```

```python
all_genres = []
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

for i in df['genres']:
  for j in i:
    all_genres.append(j['name'])

genres_pd = pd.Series(all_genres)
no_movies = genres_pd.value_counts()
```

```python
no_movies.plot(kind='bar', figsize=(10,6))
plt.title('Number of Movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```
